```python
# Salvando os antigos nomes das colunas para um uso futuro
colunas_ant = dt_test.columns

# Modificando os nomes das colunas para um padrão para se trabalhar
import unidecode

colunas = []
for nome in dt_test.columns:
    text = nome.strip().replace(' ','_').lower().replace('/','_sobre_')
    text = unidecode.unidecode(text)
    colunas.append(text)
    
    
colunas
```

```python
# Atribuindo as colunas
dt_test.columns = colunas
```

## Tratamento dos dados

```python
# Removendo o ponto do Dataframe( Cuidado para não rodar mais de uma vez)
dt_test = dt_test.apply(lambda x: x.str.replace(".",'', regex=True) if x.dtype=="object" else x)

# Não funciona
# dt_test.replace(',','.', regex=True, inplace=True)
```

```python
# Substituindo a vígula por ponto no dataframe
dt_test.replace(',','.', regex=True, inplace=True)

# Também funciona
#dt_test = dt_test.apply(lambda x: x.str.replace(",",".") if x.dtype=="object" else x)
```

```python
# Visualizando o dataset
dt_test
```

```python
# Código executado apenas uma vez!
# Colunas de texto
colunas_textuais = [dt_test["papel"],
                   dt_test["segmento"]
                   ]

# Colunas numéricas
colunas_numericas = ["p_sobre_vp",
                    "valor_de_mercado",
                    "liquidez",
                    "qtd_de_imoveis",
                    "preco_do_m2",
                    "aluguel_por_m2"]  
for coluna in colunas_numericas:
    dt_test[coluna] = dt_test[coluna].astype(float)


# Colunas percentuais
colunas_percentuais = ["ffo_yield",
                      "dividend_yield",
                      "cap_rate",
                      "vacancia_media"]

for coluna in colunas_percentuais:
    dt_test[coluna] = dt_test[coluna].str.replace("%","", regex=True).astype(float)
    dt_test[coluna] = dt_test[coluna].apply(lambda x: x/100)

# Caso especial
# dt_test["cotacao"]
```

```python
# Tratando os danos NaN para um valor conhecido
dt_test.segmento = dt_test.segmento.fillna('Desconhecido') 
```

```python
# Tratando a coluna de cotação
# Quando foi feita a raspagem de dados muitos valors perderam seu separador de casa decimal
# Contudo, sabemos que todos os preços possuiam duas casas décimais, logo, podemos fazer o devido tratamento para essa coluna

# Removendo os pontos para o tratamento geral
dt_test.cotacao = dt_test.cotacao.str.replace(".","", regex=True)

# Aplicando o tratamento
dt_test.cotacao = dt_test.cotacao.astype(float).apply(lambda x: x/100)
```
