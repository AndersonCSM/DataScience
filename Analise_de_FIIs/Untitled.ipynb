```python
import requests
import pandas as pd
import numpy as np
```

```python
url = "https://www.fundsexplorer.com.br/ranking"
```

```python
response = requests.get(url)
if response.status_code == 200:
    df = pd.read_html(response.content, encoding='utf-8')[0]
```

```python
df.sort_values('Fundos', inplace=True)
```
