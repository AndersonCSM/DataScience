## 2º Tratamentos iniciais no Dataframe
Pela rápida visualização é preciso realizar alguns tratamentos básicos, como:
- remover as linhas que não são usadas para armazenar dados;
- Atualizar as colunas para nomes válidos;
- Atualizar os índices das linhas.

```python
# 1: Removendo as linhas que não armazenam dados
df.drop([i for i in range(5)], axis=0, inplace=True)
```

```python
# 2.1: Criando um array com os valores das colunas a partir do df
colunas = df.iloc[0].unique()
colunas = colunas.astype("str")
colunas = np.char.lower(colunas)
colunas = np.char.replace(colunas, " ", "_")
colunas
```

```python
# 2.2: Atualizando o Dataframe de forma a remover o elemento do índice 0, que se trata dos valores para as colunas
df = df.iloc[1::]
```

```python
# 2.3: Atualizando as Colunas
df.columns = colunas
df
```

```python
# 3: Resetando os índices das linhas
df.reset_index(drop=True, inplace=True)
df
```

## 4 tratamento de dados
A partir da nova observação, é possível identificar algumas características acerca dos dados:
- A tipagem dos dados está incoerente com seus valores, sendo necessário uma conversão do tipo de dado.
- Não existem valores inválidos(Nulos, NaN).

Logo, é preciso realizar a conversão dos tipos de dados da seguinte maneira:
- converter os valores das colunas de concurso e bola_1 até bola_6 em tipo inteiro;
- convertar os valores da coluna data em um tipo adequado para passagem de tempo.

```python
# 1: Convertendo as colunas para o tipo númerico
indice = df.columns.get_loc("bola_1") # Obtendo o índice da coluna bola_1
colunas = df.columns[indice:indice+6].tolist() # Realizando o fatiamento de listas para obter as outrsa colunas até bola_6
colunas.append("concurso") # Adicionando a coluna concurso a lista de colunas

# Convertendo as colunas para os tipos númericos
for col in colunas:
    df[col] = pd.to_numeric(df[col])
```

```python
# Observando se a conversão está correta
df.dtypes
```

```python
# 2: convertendo o tipo da coluna data
df["data"] = pd.to_datetime(df["data"], format="%d/%m/%Y")
```

```python
# Observando se a conversão está correta
df.dtypes
```

## 4º Análise dos números

Começando pela análise dos números, será realizado algumas investigações para responder a seguinte pergunta:
- Quais números são mais sorteados para cada bola nos jogos que venceram?

```python
# 1: números mais sorteados para cada bola

# Vamos avaliar para um caso e generalizar para o todo
series_1 = df["bola_1"].value_counts().sort_index()
series1 = series_1[series_1 >= 50] # Se o número tem ocorrência maior que 50 vezes é um bom número

solução = {}
solução["bola_1"] = series1
solução
```

```python
# generalizando para todas as bolas

indice = df.columns.get_loc("bola_1")
colunas = df.columns[indice:indice+6].tolist()

solution = {}
for col in colunas:
    series = df[col].value_counts().sort_index()
    series = series[series >= 50]
    
    solution[col] = series

solution = pd.DataFrame(solution)
solution.fillna(0)  
```

```python
# Obtendo uma solução gráfica

fig = plt.figure(figsize=(11,17))

for col, i in zip(colunas, range(1,7)):
    serie = solution.loc[solution[col] > 0, col]
    plt.subplot(3,2,i)
    serie.plot.bar(y=serie, width=0.8)
    
    plt.tick_params(axis='x', labelrotation=0)
    plt.xlabel("Número sorteados")
    plt.ylabel("Ocorrências")
    plt.title(f"Números mais sorteados para a {col}")
    plt.grid(alpha=0.2)
    

plt.subplots_adjust(hspace=0.5)
fig.set_size_inches(9, 15)
plt.show()
```

Agora é possível observar quais são os números que foram mais sorteados para cada bola.

- Bola 1: número 28;
- Bola 2: número 5;
- Bola 3: número 27;
- Bola 4: número 37;
- Bola 5: número 35;
- Bola 6: número 23;

Relembrando a segunda pergunta:
- Quais números mais sairam nos jogos? 

```python
# Primeiro vamos contar a frequência de cada número por coluna
indice = df.columns.get_loc("bola_1")
colunas = df.columns[indice:indice+6].tolist()
numeros = df[colunas].apply(pd.Series.value_counts)

# O que estamos sendo mostrado é a frequência de cada número de 1 a 60(representado pelo index) para cada coluna(representado pelas colunas bolas)
numeros
```

```python
# Somando a frequência de cada número é possível obter a frequência total de cada número para todos os jogos
sum_numeros = numeros.sum(axis=1)

# Usando o método nlargest é possível descobrir os 6 números que mais saem

top6 = sum_numeros.nlargest(6).index.tolist()
top6
```

Assim, é possível notar que os números vencedores que mais saem são: 53, 10, 5, 33, 37, 4

## 5º Preparando a base de dados

### Parte I

Para prosseguir no estudo de machine learning é preciso ajustar a base de dados de acordo com o modelo de aprendizado. Para o problema de identificar o próximo conjunto de números vencedores na Mega Sena, é preciso descobrir quais são as características e qual é a variável alvo, abaixo é possível visualizar perguntas que facilitam essa visualização:
- Os números sorteados são ganhadores? se sim, verdadeiro, se não, falso
- A ordem dos números importa? não
- Quais as características dos jogos? os números jogados

Para solucionar as questões será preciso realizar uma série de transformações nos dados, além de gerar números sorteados que não venceram para que o algoritmo reconheça um jogo que venceu e um jogo que perdeu, também é preciso associar uma característica a um jogo vencedor e a um jogo perdedor.

Trabalhando por partes, é possível gerar um conjunto de todas as possibilidade de jogos, mas por se tratar de um problema com uma enorme quantidade de dados, nossos códigos não irão funcionar tão bem, sendo interessante pegar apenas uma amostra dos possíveis jogos.

```python
# Gerando todos as possibilidades de jogadas
import itertools

# Definindo o número total de bolas e o número de bolas em cada jogo
num_bolas = 60
num_escolhas = 6

# Criando uma lista de números de 1 a 60
bolas = list(range(1, num_bolas+1))

# Usando a biblioteca itertools para gerar todas as combinações possíveis de 6 bolas
comb = itertools.combinations(bolas, num_escolhas)

# Armazenando cada combinação em uma lista de jogos
jogos = np.array(list(comb)).reshape(-1, num_escolhas)

df_jogos = pd.DataFrame(jogos, columns=["bola_1", "bola_2", "bola_3", "bola_4", "bola_5","bola_6"])
```

```python
# visualizando os jogos 
df_jogos
```

### Parte II
Agora existe tanto um conjunto de jogos totais como de jogos vencedores. É interessante ter apenas um dataset com as características(bolas sorteadas) e o alvo(se é um jogo vencedor ou não), para realizar essa tarefa é preciso realizar algumas ações:
- remover os jogos vencedores do total de jogos(evitando duplicidade);
- sortear uma quantidade de jogos que não venceram e unir em um dataset final;
- adicionar a variável target aos jogos, se o jogo for vencendor 1, senão 0.

```python
# 1- Ordenando os números dos jogos vencedores e transformando em um array
df_win = df.loc[: , ["bola_1", "bola_2", "bola_3", "bola_4", "bola_5","bola_6"]]
srr_win = df_win.values
srr_win.sort()
srr_win.shape
```

```python
# visualizando o array
srr_win
```

```python
# pegando uma amostra do total de jogos de mesmo tamanho dos jogos vencedores e salvando em um array
df_new_jogos = df_jogos.sample(2581)
srr_loss = df_new_jogos.values
srr_loss
```

```python
# verificando se existem números repetidos
numeros_repetidos = (np.where(np.all(srr_loss == srr_win, axis=1)))
numeros_repetidos
```

```python
#  transformando os arrays de jogos vencedores em um dataframe e adicionando a coluna target
dt_win = pd.DataFrame(srr_win, columns=["bola_1", "bola_2", "bola_3", "bola_4", "bola_5","bola_6"])
dt_win["y"] = 1

#  transformando os arrays de jogos perdedores em um dataframe e adicionando a coluna target
dt_loss = pd.DataFrame(srr_loss, columns=["bola_1", "bola_2", "bola_3", "bola_4", "bola_5","bola_6"])
dt_loss["y"] = 0

# unindo os dois dataframes
dt_final = pd.concat([dt_loss, dt_win])
```

```python
# visualizando o dataframe final
dt_final
```

## 6º base de dados finais e pré-testes

Com o conjunto de dados prontos, é possível realizar os primeiros passos na modelagem da rede, os seguintes passos serão realizados:
- Definir as características;
- Definir o target;
- Definir conjunto de treino e de teste;
- Definir um modelo de aprendizado coerente com o problema: abordaremos dois modelos: o de regressão logística e o de uma rede neural;
- predição do próximo valor.

```python
# importando as bibliotecas
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LogisticRegression
```

```python
# definindo as características e a variável target
y = dt_final["y"]
X = dt_final.drop(["y"], axis=1)
```

```python
# treinando um modelo de regressão logística
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

modelo = LogisticRegression()
modelo.fit(X_train, y_train)
```

```python
# validando o modelo com base na acurácia
y_pred = modelo.predict(X_test)
acuracia = accuracy_score(y_test, y_pred)
print('Acurácia: ', acuracia)
```

```python
# treinando um modelo para redes neurais
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from keras.models import Sequential
from keras.layers import Dense


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Cria o modelo de rede neural
modelo = Sequential()
modelo.add(Dense(64, input_dim=6, activation='relu'))
modelo.add(Dense(1, activation='sigmoid'))

# Compila o modelo
modelo.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Treina o modelo
modelo.fit(X_train, y_train, epochs=1000, batch_size=32)

# Faz previsões no conjunto de teste
y_pred = modelo.predict(X_test)
y_pred = [1 if pred > 0.5 else 0 for pred in y_pred]

# Avalia a acurácia do modelo
acuracia = accuracy_score(y_test, y_pred)
print('Acurácia: ', acuracia)
```
